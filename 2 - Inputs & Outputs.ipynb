```python
                  ##########   #####   Input() function    #######   ########
    
 =>  input() is used in Python for taking inputs from console given by user. 

 =>  Whatever we enters as an input, input() converts it in a string and store them in a string.
 
 =>  When input() function executes itself, program flow will be stopped until user gives an input.

  =>  Whatever you enter as input, input() function convert it into a string. If you enter an integer value still
      input() function will convert it into a string. You need to explicitly convert it into an integer in 
       your code using typecasting.
```

```python
# Python program to demonstrate the working of input() function

var = input("Enter any value : ")

print ("The value stored in the variable 'var' is :", var)
print (type(var))       
```

```python
  ###########   ##########    Taking multiple inputs at a time using split() method    ##########   ############

 =>  split() is used to take multiple inputs at a time in python.
  
 =>  It splits the input by using a separator. If separator is not given, white-space is used as  separator.

   >>>  # Taking three inputs at a time 
   >>>  x, y, z = input("Enter 3 values : ").split() 

   >>>  # Printing the variables using format() method
   >>>  print("First number is {} and second number is {}, 3rd no. is {}".format(x, y, z)) 
```

```python
# Creating a List by taking multiple values at a time and typecasting them into integer

x = list(map(int, input("Enter a multiple value: ").split())) 

print("List of students: ", x) 

print(type(x))
```

## <b> Taking multiple inputs at a time using List-comprehension method </b>

```python
# Taking two input at a time 

x, y = [int(x) for x in input("Enter two value: ").split()] 

x
```

```python
# Taking multiple inputs at a time  

x = [int(x) for x in input("Enter multiple value: ").split()] 

type(x)
```

##  Outputs in Python 

```python
        #################      ##############    Print() function     ##############      ################ 
    
 =>  print() is used to display the outputs in Python language.
 
     >>>  print( value(s), sep= '', end = '', file, flush)

      
  Parameters:
        
      *  value(s) ->  This specifies how many values are going to be be passed. All values Will be converted to a
                        string before get printed.
        
      *  sep or separator (optional) ->  This specifies how to separate the multiple objects passing 
                                           through print().  Default value : 'space'
       
      *  end (Optional) ->  This specifies what to print at the end. Default value : '\n'
        
      *  file (Optional) ->  An object with a write method. Default :sys.stdout
        
      *  flush (Optional) ->  A Boolean, specifying if the output is flushed (True) or buffered (False).
                                 Default: False
```

```python
#  Python code to demonstrate the print() function

print ("Hello", "World!", "Hare", "Krishna", "is", 'a', "stupid", "boy.", sep=" ", end=" ")
```

```python
#  'end' parameter of print() function used in python....

 =>  By default this 'end' have a value of "\n", which results to print the next character into newline.
 
 =>  'end' can take value of -->  ' ', ',', '@', '/', '_', '-'  etc.
```

```python
# 'sep' parameter of print() function used in python.....

 =>  'sep' means 'separator'
 
 =>  By default, 'sep' has a value of a single space ' ', 
 
 =>  We can give any value to this parameter like as -->  '@', ',', '.', '/', '_', '-'  etc. .
```

```python
# Python program to show the use of sep parameter

print ('Hello', 'world!', sep="@")
print ('Hello', 'world!', sep="_")
print ('Hello', 'world!', sep="-")
```

```python
#  ##########   ###########  Printing Multiple variables in python  ############   ##########  ############
    
 >>> print(1)         -->   1           #  THis is an integer.

 >>> print((1))      --->   1           #  This is an integer
    
 >>> print(1, 2)     --->   1 2         #  2 'int' values are passing

 >>> print((1, 2))   --->   (1, 2)      #  a tuple is passing
```
